# Elexon API

```python
import elexon_client
import pandas as pd
```

## Single day

```python
params = {
    'ServiceCode': 'B1760',
    'SettlementDate': pd.datetime(2018, 6,15),
    'Period': '*'
}

elexon_client.query_API(**params).tail()
```

```python
params = {
    'ServiceCode': 'B1030',
    'StartDate': pd.datetime(2018,1,1),
    'EndDate': pd.datetime(2018,9,1),
    'StartTime': pd.datetime(2018,1,1,1),
    'EndTime': pd.datetime(2018,1,3,23)
}

elexon_client.query_API(**params).tail()
```

## Multiple days

```python
params = {
    'ServiceCode': 'B1430',
    'Period': '*'
}

start = pd.datetime(2018,1,1)
end   = pd.datetime(2018,1,3)
date_param = 'SettlementDate'

df = elexon_client.query_multiple_days(start, end, date_param, **params)
df.tail()
```
