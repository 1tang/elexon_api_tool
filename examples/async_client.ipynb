# Elexon Async Client

```python
from elexon_api import AsyncClient as Client
from elexon_api import extract_df

import pandas as pd
```

### normal query

```python
client = Client.from_key_file()

service_code = "DYNBMDATA"

params  = {
    'SettlementDate': pd.datetime(2019,1,1),
    'SettlementPeriod': 1
}
```

```python
r = await client.query(service_code, **params)
df = extract_df(r)
```

```python
df.head()
```

### multiple asynce queries

```python
import asyncio
```

```python
service_code = 'BOD'
date = pd.datetime(2018,1,1)
```

```python
task_l = []
for period in range(1, 48 + 1):

    params = {
        'SettlementDate': date,
        'SettlementPeriod': period}

    task_l.append(client.query(service_code, check_query=False, **params))

r = await asyncio.gather(*task_l)
```

```python
df = pd.concat(map(extract_df, r), axis=0)
```
